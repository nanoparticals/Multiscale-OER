```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
K_B = 1.38e-23  # Boltzmann constant (J/K)  
H_PLANCK = 6.626e-34  # Planck constant (J·s)  
T = 298.15  # Temperature (K)  
A = 1e9  # Pre-exponential factor (unit varies with reaction order)  

# Hypothetical initial Gibbs free energy values (unit: eV)  
G_values = [0, -0.11, 1.46, 2.81, 4.92]  

Ea0 = 0.26  # Base activation energy (eV)  
beta = 0.5  # Symmetry factor (unitless)  
pe_g = 1  # Partial pressure of product (arbitrary unit)  
H2O_g = 1  # Partial pressure of water vapor (arbitrary unit)  
O2_g = 2.34e-5  # Partial pressure of oxygen (arbitrary unit)  
F = 96485.33212  # Faraday constant (C/mol)  
```

```python
def calculate_modified_G(G_values, V):
    # The first term remains unchanged, subsequent terms subtract multiples of voltage
    modified_G_values = [G - i * V for i, G in enumerate(G_values)]
    # Note: The first G value should not be modified, so we restore the original value
    modified_G_values[0] = G_values[0]
    return modified_G_values
```

```python
def calculate_Ea(Ea0, beta, delta_G):
    """
    Calculate activation energy (Ea) based on initial activation energy, beta parameter, and Gibbs free energy.

    Parameters:
    - Ea0: Initial activation energy (in eV or kJ/mol, must match delta_G units)
    - beta: Symmetry factor (unitless, typically 0 < beta < 1 in Marcus theory)
    - delta_G: Gibbs free energy change of reaction (same units as Ea0)

    Returns:
    - Ea: Calculated activation energy (same units as inputs)

    Calculation Rules:
    1. Returns 0.224 if -beta*delta_G > Ea0 (activationless regime)
    2. Returns delta_G + 0.224 if (1-beta)*delta_G > Ea0 (inverted regime)
    3. Otherwise returns Ea0 + beta*delta_G + 0.224 (normal Marcus regime)
    """
    if -beta * delta_G > Ea0:
        return 0.224  # Activationless limit
    elif (1 - beta) * delta_G > Ea0:
        return delta_G + 0.224  # Inverted regime
    else:
        return Ea0 + beta * delta_G + 0.224  # Normal Marcus behavior
```

```python
# Calculates forward and reverse reaction rate constants (k⁺ and k⁻) 
def calc_k(Ea, delta_G, V, T=T):
    Ea_reverse = Ea - delta_G
    kplus = A * np.exp(-Ea*1.60218e-19 / (K_B * T))
    kminus = A * np.exp(-Ea_reverse*1.60218e-19 / (K_B * T))
    return kplus, kminus
```

```python
# ODEs
def odes(theta, t, kps, kms, pe_g, H2O_g, O2_g):
    theta_1, theta_oh, theta_o, theta_ooh = theta
    r1 = kps[0] * theta_1 * H2O_g - kms[0] * theta_oh * pe_g
    r2 = kps[1] * theta_oh - kms[1] * theta_o * pe_g
    r3 = kps[2] * theta_o * H2O_g - kms[2] * theta_ooh * pe_g
    r4 = kps[3] * theta_ooh - kms[3] * theta_1 * pe_g * O2_g
    dtheta_1_dt = -r1 + r4
    dtheta_oh_dt = r1 - r2
    dtheta_o_dt = r2 - r3
    dtheta_ooh_dt = r3 - r4
    return [dtheta_1_dt, dtheta_oh_dt, dtheta_o_dt, dtheta_ooh_dt]
    

```

```python
initial_theta = [1, 0, 0, 0]  # Initial coverage fractions

voltage_range = np.linspace(1.2, 2, 101)  # Voltage sweep from 1.2V to 2V (101 steps)

time_range = np.linspace(0, 1, 100)  # Time span from 0 to 1 (arbitrary units or seconds)
```

```python

theta_time_results = []  # Coverage vs. time
theta_voltage_results = []  # Coverage vs. voltage
```

```python
# Simulation Process
for V in voltage_range:
    #  Calculate modified Gibbs free energy
    modified_G_values = calculate_modified_G(G_values, V)
    print(modified_G_values)  

    # Calculate delta_G (difference between adjacent modified G values) 
    delta_Gs = [modified_G_values[i+1] - modified_G_values[i] for i in range(len(modified_G_values) - 1)]

    # Calculate activation energy
    Eas = [calculate_Ea(Ea0, beta, delta_G) for delta_G in delta_Gs]

    # Calculate rate constants  
    kps = [calc_k(Ea, delta_G, V)[0] for Ea, delta_G in zip(Eas, delta_Gs)]
    kms = [calc_k(Ea, delta_G, V)[1] for Ea, delta_G in zip(Eas, delta_Gs)]

    # Output forward and reverse rate constants for each voltage  
    print(f"Rate constants at {V:.2f} :")
    for i, (kp, km) in enumerate(zip(kps, kms)):
        print(f"Reaction step {i+1}: k+ = {kp:.2e} /s, k- = {km:.2e} /s")
    print("-" * 30)

    # Solve the system of differential equations 
    theta_t = odeint(odes, initial_theta, time_range, args=(kps, kms, pe_g, H2O_g, O2_g))
    theta_time_results.append(theta_t)
    theta_voltage_results.append(theta_t[-1])
 
```

```python
plt.figure(figsize=(14, 7))
for i, V in enumerate(voltage_range):
    plt.plot(time_range, theta_time_results[i][:, 0], label=f'V={V:.1f}V')
plt.xlabel('Time (s)')
plt.ylabel('Coverage of H2O')
plt.title('Coverage of H2O Over Time at Different Voltages')
plt.legend()
plt.show()
```

```python
theta_voltage_results = np.array(theta_voltage_results)
plt.figure(figsize=(14, 7))
plt.plot(voltage_range, theta_voltage_results[:, 0], label='Theta (H2O)')
plt.plot(voltage_range, theta_voltage_results[:, 1], label='Theta_OH')
plt.plot(voltage_range, theta_voltage_results[:, 2], label='Theta_O')
plt.plot(voltage_range, theta_voltage_results[:, 3], label='Theta_OOH')
plt.xlabel('Voltage (V)')
plt.ylabel('Coverage at Final Time')
plt.title('Coverage of Species vs Voltage')
plt.legend()
plt.show()
```
